```python
# reformatting the data by changing everything to lowercase and replacing multiple spaces and dash with an underscore
orig_columns = list(data.columns)
new_columns = []
for col in orig_columns:
    new_columns.append(col.strip().replace(' ',' ').replace(' ','_').replace('-','_').lower())
data.columns = new_columns
```

```python
data.drop(columns=['country','year','status','percentage_expenditure','bmi'
                   ,'under_five_deaths','total_expenditure','gdp',
                   'income_composition_of_resources'],inplace=True)
```

```python
data.head(2)
```

```python
data.drop(columns=['population','thinness__1_19_years','thinness_5_9_years'
                   ,'infant_deaths','schooling','adult_mortality'],inplace=True)
```

```python
data.head(2)
```

```python
data.isna().sum()
```

```python
data['life_expectancy'].fillna(data['life_expectancy'].median(),inplace=True)

data['alcohol'].fillna(data['alcohol'].median(),inplace=True)

data['hepatitis_b'].fillna(data['hepatitis_b'].median(),inplace=True)

data['polio'].fillna(data['polio'].median(),inplace=True)

data['diphtheria'].fillna(data['diphtheria'].median(),inplace=True)
```
